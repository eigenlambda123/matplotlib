```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('classic')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100, c='green', 
            edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()

plt.show()
```

```python
# Multiple Colors

plt.style.use('classic')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Greens',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar() # Sidebar
cbar.set_label('Satisfaction') # Sidebar label

plt.tight_layout()
plt.show()
```

```python
# Different Sizes

# Multiple Colors

plt.style.use('classic')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='Greens',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar() # Sidebar
cbar.set_label('Satisfaction')  # Sidebar label

plt.tight_layout()

plt.show()
```

```python
# Read World Data

plt.style.use('classic')

data = pd.read_csv('youtube_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
# Read World Data (log scale)

plt.style.use('classic')

data = pd.read_csv('youtube_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar() # Sidebar
cbar.set_label('Like/Dislike Ratio') # Sidebar label

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
